# Visualización de estadística básica

```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
def readData():
    data = pd.read_csv("data1.txt", sep = "\t", header = 0)
    return data

df = readData()
```

```python
plt.boxplot(df)
plt.show()
```

```python
plt.hist(df)
plt.show()
```

```python
df.corr(method = 'pearson')
```

```python
y = df.iloc[: ,1]
x = range(len(y))
plt.scatter(x,y)
plt.show()
```

```python
ax = sb.heatmap(df)
```
